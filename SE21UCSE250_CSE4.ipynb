**Importing necessary libraries**

```python
import re
from collections import Counter
import matplotlib.pyplot as plt
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
nltk.download('wordnet')
```

```python
from google.colab import drive
file_path = "/content/drive/MyDrive/NLP class/Sherlock Holmes.txt"
```

```python
# Reading the text file
with open('Sherlock Holmes.txt', 'r', encoding='utf-8') as file:
    text = file.read()
```

**TEXT PREPROCESSING**

```python
#tokenising and normaliozing text
def preprocess_text(text):
    tokens = word_tokenize(text)
    tokens = [token.lower() for token in tokens]

    return tokens
# Filter out non-relevant words using regular expression pattern
def extract_relevant_words(tokens):
    relevant_words = [word for word in tokens if re.match(r'\b[a-z]+\b', word)]

    return relevant_words

# Initialize stemmer and lemmatizer
def stem_and_lemmatize(tokens):
    porter = PorterStemmer()
    lemmatizer = WordNetLemmatizer()

    # Stem and lemmatize tokens
    stemmed_tokens = [porter.stem(token) for token in tokens]
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

    return stemmed_tokens, lemmatized_tokens
# Preprocessing the text
tokens = preprocess_text(text)
relevant_words = extract_relevant_words(tokens)
stemmed_tokens, lemmatized_tokens = stem_and_lemmatize(relevant_words)
```

**PLOTS**

```python
def plot_word_frequency(tokens, title):
    word_freq = Counter(tokens)
    common_words = word_freq.most_common(10)
    words, freqs = zip(*common_words)

    plt.figure(figsize=(10, 6))
    plt.plot(words, freqs, marker='o', color='green', linestyle='-')
    plt.title(title)
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

# Plotting word frequency for stemmed and lemmatized words
plot_word_frequency(stemmed_tokens, 'Word Frequency (Stemmed)')
plot_word_frequency(lemmatized_tokens, 'Word Frequency (Lemmatized)')
```

List of all the tokens, stemmed, lemmatized, normalized, and extracted relevant words

```python
print("Tokens:", tokens[:10])
print("Relevant words:", relevant_words[:10])
print("Stemmed tokens:", stemmed_tokens[:10])
print("Lemmatized tokens:", lemmatized_tokens[:10])
```

**words ending with 'ing'**

```python
#words ending with 'ing' for stemmed and lemmatized tokens

def get_words_ending_with_ing(tokens):
    ing_words = [word for word in tokens if word.endswith('ing')]
    return ing_words
stemmed_ing_words = get_words_ending_with_ing(stemmed_tokens)
lemmatized_ing_words = get_words_ending_with_ing(lemmatized_tokens)

print("Stemmed words ending with 'ing':", stemmed_ing_words[:100])
print("Lemmatized words ending with 'ing':", lemmatized_ing_words[:100])
```

```python
def plot_ING_frequency(tokens, title):
    word_freq = Counter(tokens)
    common_words = word_freq.most_common(10)
    words, freqs = zip(*common_words)

    plt.figure(figsize=(10, 6))
    plt.plot(words, freqs, marker='o', color='red', linestyle='-')
    plt.title(title)
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

# Plotting word frequency for stemmed and lemmatized words
plot_ING_frequency(stemmed_ing_words, 'ING Frequency (Stemmed)')
plot_ING_frequency(lemmatized_ing_words, 'ING Frequency (Lemmatized)')
```
